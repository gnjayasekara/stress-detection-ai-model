```python
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
from keras import preprocessing
from keras.preprocessing.image import load_img
from keras.models import Sequential
from keras.layers import Dense,Conv2D, Dropout, Flatten, MaxPooling2D
import os
import pandas as pd
import numpy as np
```

```python
train_dir='images/train'
test_dir='images/test'
```

```python
def createdataframe(dir):
    image_paths = []
    labels = []
    for label in os.listdir(dir):
        for imagename in os.listdir(os.path.join(dir,label)):
            image_paths.append(os.path.join(dir,label,imagename))
            labels.append(label)
        print(label, "completed")
    return image_paths,labels
        
```

```python
train = pd.DataFrame()
train['image'], train['label'] = createdataframe(train_dir)
```

```python
print(train)
```

```python
test = pd.DataFrame()
test['image'], test['label'] = createdataframe(test_dir)
```

```python
print(test)
```

```python
from tqdm.notebook import tqdm
from PIL import Image
```

```python
def extract_features(images):
    features = []
    for image in tqdm(images):
        img = load_img(image, color_mode='grayscale', target_size=(48, 48))
        img = np.array(img)
        features.append(img)
    features = np.array(features)
    features = features.reshape(len(features),48,48,1)
    return features
```

```python
train_features = extract_features(train['image'])
```

```python
test_features = extract_features(test['image'])
```

```python
x_train = train_features/255.0
x_test = test_features/255.0
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(train['label'])
```

```python
y_train = le.transform(train['label'])
y_test = le.transform(test['label'])
```

```python
y_train = to_categorical(y_train,num_classes = 7)
y_test = to_categorical(y_test,num_classes = 7)
```

```python
model = Sequential()
# convolutional Layers
model.add(Conv2D(128,kernel_size=(3,3), activation='relu', input_shape=(48,48,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(256, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Conv2D(512, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.4))

model.add(Flatten())
# fully connected Layers
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
# output layer
model.add(Dense(7, activation='softmax'))
```

```python
model.compile(optimizer= 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x= x_train, y= y_train, batch_size = 128, epochs = 100, validation_data = (x_test,y_test))
```

```python
model_json = model.to_json()
with open("stressdetector.json",'w') as json_file:
    json_file.write(model_json)
model.save("stressdetectorV1.h5")
```

```python
from tensorflow.keras.models import load_model

# Load the pre-trained model (replace 'your_model.h5' with the actual model path)
model = load_model('stressdetectorV1.h5')

```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Directory containing test images
test_dir='images/test'
true_labels = []
predicted_labels = []

# Assuming your labels are defined in a dictionary or list
labels = ['angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised']  # Update with actual mapping

# Loop through each label directory
for label in labels:
    label_dir = os.path.join(test_dir, label)
    
    # Loop through each image in the label directory
    for filename in os.listdir(label_dir):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # Add other image formats if needed
            true_labels.append(label)
            
            # Load and preprocess the image
            img_path = os.path.join(label_dir, filename)
            img = load_img(img_path, color_mode='grayscale', target_size=(48, 48))
            img = img_to_array(img)
            img = np.expand_dims(img, axis=0)  # Add batch dimension
            
            # Predict using the model
            pred = model.predict(img)
            pred_label = labels[np.argmax(pred)]
            predicted_labels.append(pred_label)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Calculate and print the metrics
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels, average='weighted')
recall = recall_score(true_labels, predicted_labels, average='weighted')
f1 = f1_score(true_labels, predicted_labels, average='weighted')
conf_matrix = confusion_matrix(true_labels, predicted_labels, labels=labels)
class_report = classification_report(true_labels, predicted_labels, target_names=labels)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

# Compute the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=labels)

# Compute per-label accuracy
per_label_accuracy = {}
for i, label in enumerate(labels):
    true_positive = cm[i, i]  
    total_instances = np.sum(cm[i, :])  
    per_label_accuracy[label] = true_positive / total_instances if total_instances > 0 else 0.0

# Print per-label accuracy
print("Per-Label Accuracy:")
for label, accuracy in per_label_accuracy.items():
    print(f"{label}: {accuracy:.2f}")
```

```python
from sklearn.metrics import classification_report
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Compute classification report
report = classification_report(true_labels, predicted_labels, target_names=labels, output_dict=True)

# Convert the report to a DataFrame
report_df = pd.DataFrame(report).transpose()
metrics = report_df.loc[labels, ['precision', 'recall', 'f1-score']].reset_index()

# Add accuracy as a metric
metrics['accuracy'] = [per_label_accuracy[label] for label in labels]
```

```python
# Metrics Data
metrics_labels = metrics['index'].values 
precision = metrics['precision'].values
recall = metrics['recall'].values
f1_score = metrics['f1-score'].values
accuracy = metrics['accuracy'].values

# X positions for groups
x = np.arange(len(metrics_labels))

# Bar width
bar_width = 0.2

# Plotting grouped bar chart
plt.figure(figsize=(12, 6))
plt.bar(x - 1.5 * bar_width, accuracy, width=bar_width, label='Accuracy', color='skyblue')
plt.bar(x - 0.5 * bar_width, precision, width=bar_width, label='Precision', color='lightgreen')
plt.bar(x + 0.5 * bar_width, recall, width=bar_width, label='Recall', color='lightcoral')
plt.bar(x + 1.5 * bar_width, f1_score, width=bar_width, label='F1-Score', color='gold')

# Customizing plot
plt.xlabel('Labels', fontsize=12)
plt.ylabel('Metric Values', fontsize=12)
plt.title('Evaluation Metrics by Label', fontsize=16)
plt.xticks(x, metrics_labels, fontsize=10, rotation=45)
plt.ylim(0, 1.0)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add metric values on top of bars
for i in range(len(metrics_labels)):
    plt.text(x[i] - 1.5 * bar_width, accuracy[i] + 0.02, f'{accuracy[i]:.2f}', ha='center', fontsize=9)
    plt.text(x[i] - 0.5 * bar_width, precision[i] + 0.02, f'{precision[i]:.2f}', ha='center', fontsize=9)
    plt.text(x[i] + 0.5 * bar_width, recall[i] + 0.02, f'{recall[i]:.2f}', ha='center', fontsize=9)
    plt.text(x[i] + 1.5 * bar_width, f1_score[i] + 0.02, f'{f1_score[i]:.2f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
